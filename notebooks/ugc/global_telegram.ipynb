```python
parameters = [
    {'param_name': 'offset_date', 'kwargs':{'offset_date':datetime.fromisoformat('2023-10-01T00:00:00Z')}}, # get messages before this datetime
    {'param_name': 'reverse_order', 'kwargs':{'reverse': True}}, # get messages from oldest to newest
    {'param_name': 'search', 'kwargs':{'search': 'chocolate'}}, # search for messages containing the term "chocolate"
    {'param_name': 'specific_ids', 'kwargs':{'ids': [5000, 2895, 6010, 6322]}}, # get specific messages by their ID
    {'param_name': 'offset_id', 'kwargs':{'offset_id': 6000}}, # retrieve messages posted before the one with this ID
]

peer_channel = PeerChannel(channel_id=1593453357)# object that identifies a channel

async def get_consistent_data(api_client: TelegramClient, peer: PeerChannel, parameters: list[dict], limit: int = 5):
    for idx, param in enumerate(parameters):
        index = idx + 1 
        data_dir = "../../data" # SET PATH
        filename = f"br-telegram-UGC-question-21-run-{index}-{param['param_name']}-{datetime.now().strftime('%Y-%m-%dT%H-%M-%S-%f')}.json"
        
        # implement the function here
        response = await api_client.get_messages(
            entity=peer,# object for identifying a group, channel or user
            limit=limit,# maximum number of messages to retrieve
            **param['kwargs']
        )
        
        messages = [# convert some of telethon messages fields to a JSON serializable format (e.g. datetime to isoformat)
            {
                'id': m.id,
                'peer_id': m.peer_id.channel_id,
                'date': m.date.isoformat(),
                'message': m.message
            } for m in response
        ]

        with open(f"{data_dir}/{filename}", "w") as fout:
            json.dump(messages, fout)

await asyncio.create_task(get_consistent_data(api_client=api_client, peer=peer_channel, parameters=parameters))
```

**OC25: Does the platform’s API allow for filtering data by specific time periods?**

*This item verifies whether the API allows applying temporal filters to data extraction. The assessment should test the endpoint for the main content type to confirm that public post data can be filtered by custom time ranges.*


```python
min_date = datetime.fromisoformat('2025-09-29T00:00:00Z')
max_date = datetime.fromisoformat('2025-10-01T00:00:00Z')
peer_channel = PeerChannel(channel_id=1742616113)# object that identifies a channel

async def get_messages_time_range(api_client: TelegramClient, peer: PeerChannel, min_date: datetime, max_date: datetime, limit: int = 100):
    return await api_client(functions.messages.SearchRequest(
        peer=peer_channel,
        q='', # Empty string for all messages texts
        filter=InputMessagesFilterEmpty(), # Search for all message types
        # The five positional arguments required by the function:
        offset_id=0,       # Start from the beginning
        add_offset=0,      # No additional offset
        max_id=0,          # Search up to the newest message
        min_id=0,          # Search down to the oldest message
        hash=0,            # Default hash value
        min_date=min_date,
        max_date=max_date,
        limit=100
    ))

result = await asyncio.create_task(get_messages_time_range(api_client=api_client, peer=peer_channel, min_date=min_date, max_date=max_date))
result_selection = [
    {
        'channel_id': x.peer_id.channel_id,
        'date': x.date.isoformat(),
        'message_id': x.id,
        'message_type': x.to_dict()['_'],
        'message': x.message
    } for x in result.messages
]

df = pd.DataFrame(result_selection)
df.sort_values(by='date', ascending=False)
```
